```python
# check if num is palindrome
# palindrome is something which read either left to right or right to left 
n = 1234
num = n
result = 0
while num>0:
  last_digit = num%10
  result = result*10+last_digit
  num = num//10
  if result ==n:
    print("palindrome")
  else:
    print("not palindrome")
```

```python
# if the number is armstrong or not
# armstrong num is a number which is equal to the sum of its own digits raised to the power of the number of digits.
n = 153
num=n
total=0
nod = len(str(n))
while num>0:
  last_digit = num%10
  total +=last_digit**nod
  num = num//10
  print(total)
  if total == n:
    print("armstrong")
  else: 
    print("not armstrong")
```
